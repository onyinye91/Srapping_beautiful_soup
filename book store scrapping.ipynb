```python
import requests
```

```python
url = 'https://www.bookdepository.com/search?searchTerm=python&search=Find+book'
```

```python
r = requests.get(url)
```

```python
print(r.text[:100])
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
results = soup.findAll('', attrs = {'class': 'item-info'})
len(results)
```

```python
for name in soup.find_all('',{'itemprop':'name'}):
    print(name.get('content'))
```

```python
result_one = results[0]
result_one
```

```python
#name_of_book = result_one.find('h3').text[22:-1]
#name_of_book
#results.find('h3').text[22:-1]
new_item = []
data_list = []
for item in soup.findAll('',{'itemprop':'name'}):
    new_item.append(item.get('content'))
data_list.append(new_item)
```

```python
#author = result_one.find('span').text[2:-2]
#author
new_item = []
for item in soup.findAll('',{'itemprop',"contributor"}):
    new_item.append(item.get('content'))
data_list.append(new_item)
```

```python
#date = result_one.find('p', attrs = {'class':'published'}).text[20:]
#date
new_item = []
for item in soup.findAll('',{'class':'published'}):
    new_item.append(str(item)[-15:-4])
data_list.append(new_item) 
```
